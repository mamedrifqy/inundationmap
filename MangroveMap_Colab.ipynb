## Cell 4 — Write server code to disk
⚠️ **Always run this cell after restarting the runtime.**

```python
import base64, os, sys

# Remove any old cached version
for k in list(sys.modules.keys()):
    if "main" in k: del sys.modules[k]

os.makedirs("mangrove_app/backend", exist_ok=True)
_code = base64.b64decode("").decode("utf-8")
with open("mangrove_app/backend/main.py", "w") as f:
    f.write(_code)

print(f"✓ main.py written ({len(_code):,} chars)")
print("Last 3 lines:")
print("\n".join(_code.strip().splitlines()[-3:]))
```

## Cell 5 — Start the server 🚀
Run this to start (or restart) the server. Copy the URL it prints.

```python
import nest_asyncio, sys, os, time, threading
import uvicorn
from pyngrok import ngrok, conf
nest_asyncio.apply()

# Kill old ngrok
try: ngrok.kill()
except: pass
time.sleep(2)

# Clear module cache
for k in list(sys.modules.keys()):
    if "main" in k: del sys.modules[k]

# Add backend to path
backend = os.path.abspath("mangrove_app/backend")
if backend not in sys.path: sys.path.insert(0, backend)
os.environ["GEE_PROJECT"] = GEE_PROJECT

# Start FastAPI server
threading.Thread(
    target=lambda: uvicorn.run("main:app", host="0.0.0.0", port=8000, log_level="info"),
    daemon=True
).start()
time.sleep(4)

# Start ngrok tunnel
conf.get_default().auth_token = NGROK_TOKEN
tunnel = ngrok.connect(8000, "http")
public_url = tunnel.public_url
print()
print("=" * 60)
print("  InundationMap V2 is LIVE")
print(f"  URL  : {public_url}")
print(f"  Code : M4CR2026")
print("=" * 60)
print()
print("→ Paste the URL above into the API SERVER field in the app")
```

## Cell 6 — Test server (optional but recommended)
Run this to confirm the server is working correctly before using the app.

```python
import requests
H = {"ngrok-skip-browser-warning": "true"}

# Test 1: health
h = requests.get(f"{public_url}/health", headers=H).json()
print("Health:", h)

# Test 2: list polygons
p = requests.get(f"{public_url}/polygons", headers=H,
                 params={"asset_path": ASSET_PATH}).json()
print(f"Polygons loaded: {p.get('count', 0)}")
print("First 5:")
for poly in p.get("polygons", [])[:5]:
    kel = poly["properties"].get("Kelompok") or poly["properties"].get("kelompok","?")
    print(f"  id={poly['id']:25s}  kelompok={kel:30s}  {poly['area_ha']} ha")
```
